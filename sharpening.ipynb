```python
import cv2
from skimage import data
import matplotlib.pyplot as plt
import numpy as np

image  = cv2.imread('./test/source_rgb_512.bmp')
source_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
print(image.shape)
kernel = np.array([
                [ 0, -1,  0],
                [-1,  5, -1],
                [ 0, -1,  0]
                ])


sharpened = cv2.filter2D(image, -1, kernel)
sharpened = cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB) 
sharpened_FPGA  = cv2.imread('./test/sharpened_fpga_rgb_512.bmp')
sharpened_FPGA = cv2.cvtColor(sharpened_FPGA, cv2.COLOR_BGR2RGB) 
```

```python

fi1, axs1 = plt.subplots(1, 3, figsize=(25, 10))

ax = axs1.ravel()
ax[0].imshow(source_image)
ax[0].set_title('Source image')
ax[1].imshow(sharpened)
ax[1].set_title('Sharpened image via Opencv')
ax[2].imshow(sharpened_FPGA)
ax[2].set_title('Sharpened image via FPGA')
```
